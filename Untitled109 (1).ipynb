```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
weights = {
    "f": (0.5, 0.1, 0),
    "i": (0.6, 0.2, 0),
    "c": (0.7, 0.3, 0),
    "o": (0.8, 0.4, 0)
}

Wy = 4
by = 0

X = [1,2,3,4]

h_prev = 0
C_prev = 0

print("LSTM Implementation")

for t in range(len(X)):

    x_t = X[t]
    print(f"\nTime Step t={t+1}, Input x_t={x_t}")

    Wf, Whf, bf = weights["f"]
    f_t = sigmoid(Wf*x_t + Whf*h_prev + bf)
    print(f"Forget Gate (f_t): {f_t:.3f}")

    Wi, Whi, bi = weights["i"]
    i_t = sigmoid(Wi*x_t + Whi*h_prev + bi)
    print(f"Input Gate (i_t): {i_t:.3f}")

    Wc, Whc, bc = weights["c"]
    C_tilde = tanh(Wc*x_t + Whc*h_prev + bc)
    print(f"Candidate Cell State: {C_tilde:.3f}")

    C_t = f_t*C_prev + i_t*C_tilde
    print(f"Cell State (C_t): {C_t:.3f}")

    Wo, Who, bo = weights["o"]
    o_t = sigmoid(Wo*x_t + Who*h_prev + bo)
    print(f"Output Gate (o_t): {o_t:.3f}")

    h_t = o_t * tanh(C_t)
    print(f"Hidden State (h_t): {h_t:.3f}")

    h_prev = h_t
    C_prev = C_t

y_hat = Wy*h_prev + by

print("\n--- Final Prediction ---")
print(f"Predicted next value (y_hat): {y_hat:.3f}")
```
